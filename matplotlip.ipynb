```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.xlabel("X AXIES------->")
plt.ylabel("Y AXIES------->")
plt.plot(x,y1,color='r',linestyle='-',linewidth=5)
plt.plot(x,y2,color='y',linestyle=':',linewidth=5)
plt.grid(True)
plt.title("MY FIRST MATPLOTLIB CODE")
plt.show()
```
